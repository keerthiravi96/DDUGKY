### Type casting with pandas

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('Car_sales.csv')
data.sample(10)
```

### Drop na

```python
data.dropna(inplace=True)
data.info()
```

### Checking the contents in latest launch column

```python
data.Latest_Launch.unique()
```

### Change its dtype into datetime

```python
data['Latest_Launch']=pd.to_datetime(data.Latest_Launch)
```

```python
data.head(10)
```

### Check the dtype of it:

```python
data.info()
```

### We can see that it have changed to datetime 64[ns] format

### Extracting year,month,date from the Latest_launch column

```python
data['Year']=pd.DatetimeIndex(data.Latest_Launch).year
```

```python
data['month']=pd.DatetimeIndex(data.Latest_Launch).month
```

```python
data['day']=pd.DatetimeIndex(data.Latest_Launch).day
```

```python
data.head(10)
```

### Lets convert the data type of Fuel efficiency into integer

```python
data.Fuel_efficiency=data.Fuel_efficiency.astype(int)
data.info()
```

### Converting the numerical column into categorical column

#### Lets convert the month numbers to their corresponding names

```python
data.month.replace({
    1:'January',
    2:'February',
    3:'March',
    4:'April',
    5:'May',
    6:'June',
    7:'July',
    8:'August',
    9:'September',
    10:'October',
    11:'November',
    12:'December',
},inplace=True)
```

```python
data.sample(10)
```

### Label encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data.sample(10)
```
